```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('/content/drive/MyDrive/Jaguar-01-01.webp')
img2 = cv2.imread('/content/drive/MyDrive/Keel-Billed-Toucan.webp')
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(gray1, cmap='gray')
plt.title("Grayscale Image 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gray2, cmap='gray')
plt.title("Grayscale Image 2")
plt.axis('off')
plt.show()

```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.hist(gray1.ravel(), bins=256, range=(0,256))
plt.title("Histogram Image 1")

plt.subplot(1,2,2)
plt.hist(gray2.ravel(), bins=256, range=(0,256))
plt.title("Histogram Image 2")

plt.show()


```

```python
resized1 = cv2.resize(gray1, (256,256))
resized2 = cv2.resize(gray2, (256,256))

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(resized1, cmap='gray')
plt.title("Resized Image 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(resized2, cmap='gray')
plt.title("Resized Image 2")
plt.axis('off')
plt.show()

```

```python
edges1 = cv2.Canny(resized1, 100, 200)
edges2 = cv2.Canny(resized2, 100, 200)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(edges1, cmap='gray')
plt.title("Edges Image 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edges2, cmap='gray')
plt.title("Edges Image 2")
plt.axis('off')
plt.show()
```

```python
mean = 0
std = 40

noise1 = np.random.normal(mean, std, resized1.shape)
noise2 = np.random.normal(mean, std, resized2.shape)

noisy1 = resized1 + noise1
noisy2 = resized2 + noise2

noisy1 = np.clip(noisy1, 0, 255).astype(np.uint8)
noisy2 = np.clip(noisy2, 0, 255).astype(np.uint8)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(noisy1, cmap='gray')
plt.title("Noisy Image 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(noisy2, cmap='gray')
plt.title("Noisy Image 2")
plt.axis('off')
plt.show()
```

```python
denoised1 = cv2.GaussianBlur(noisy1, (5,5), 0)
denoised2 = cv2.GaussianBlur(noisy2, (5,5), 0)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(denoised1, cmap='gray')
plt.title("Denoised Image 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(denoised2, cmap='gray')
plt.title("Denoised Image 2")
plt.axis('off')
plt.show()
```
